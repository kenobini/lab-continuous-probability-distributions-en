## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform(bottom, ceiling, count):
    return uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
```

```python
data1 = generate_uniform(bottom=10, ceiling=15, count=100)
data2 = generate_uniform(bottom=10, ceiling=60, count=1000)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, color='blue', alpha=0.7, edgecolor='black')
plt.title('Uniform Distribution (10 to 15, n=100)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title('Uniform Distribution (10 to 60, n=1000)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The first distribution (10 to 15) is more compact since the range is smaller and contains fewer points.
The second distribution (10 to 60) spans a much larger range and has more data points, making it appear smoother.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# Function to generate normally distributed random numbers
def generate_normal(mean, std_dev, count):
    return norm.rvs(loc=mean, scale=std_dev, size=count)

# Generate data
normal_data1 = generate_normal(mean=10, std_dev=1, count=1000)
normal_data2 = generate_normal(mean=10, std_dev=50, count=1000)

# Plot the histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(normal_data1, bins=50, color='purple', alpha=0.7, edgecolor='black')
plt.title('Normal Distribution (mean=10, std_dev=1)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(normal_data2, bins=50, color='orange', alpha=0.7, edgecolor='black')
plt.title('Normal Distribution (mean=10, std_dev=50)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The first distribution has most of its values concentrated near the mean (10) due to a smaller standard deviation (1).
The second distribution is much more spread out, with values widely scattered around the mean because of the larger standard deviation (50).

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# Function to generate exponentially distributed random numbers
def generate_exponential(mean, count):
    return np.random.exponential(scale=mean, size=count)

# Generate data
exp_data1 = generate_exponential(mean=10, count=10)
exp_data2 = generate_exponential(mean=10, count=100)

# Plot the histograms
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(exp_data1, bins=10, color='pink', alpha=0.7, edgecolor='black')
plt.title('Exponential Distribution (mean=10, n=10)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(exp_data2, bins=50, color='red', alpha=0.7, edgecolor='black')
plt.title('Exponential Distribution (mean=10, n=100)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The first distribution (n=10) is more jagged due to fewer samples.
The second distribution (n=100) is smoother and better represents the exponential shape.

The mean changes, so the distribution changes as well.
